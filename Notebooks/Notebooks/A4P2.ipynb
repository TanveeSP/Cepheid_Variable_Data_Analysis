```python
import pandas as pd
import numpy as np
from astropy.timeseries import LombScargle

# Load the CSV file
file_path = 'curves.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Print column names to check for correctness
print("Columns in the dataset:", data.columns)

expected_columns = ['ID', 'JD', 'apparent_magnitude']
column_mapping = {
    'id': 'ID',
    'julian_date': 'JD',
    'm': 'apparent_magnitude'  # Adjust based on actual column names
}
if not all(col in data.columns for col in expected_columns):
    print("Renaming columns to match expected names...")
    data.rename(columns=column_mapping, inplace=True)

print("Updated Columns:", data.columns)

# Dictionary
cepheid_periods = {}

# Group data by Cepheid ID
grouped = data.groupby('ID')

#  calculate the period of each cepheid
for cepheid_id, group in grouped:
    jd = group['JD'].values  # Julian Dates
    magnitude = group['apparent_magnitude'].values  # Apparent magnitudes

    # Calculating the Lomb-Scargle periodogram
    frequency, power = LombScargle(jd, magnitude).autopower()

    # Find the period corresponding to the maximum power
    best_frequency = frequency[np.argmax(power)]
    period = 1 / best_frequency

    # Store the period in the dictionary
    cepheid_periods[cepheid_id] = period

# Print the results
for cepheid_id, period in cepheid_periods.items():
    print(f"Cepheid ID: {cepheid_id}, Period: {period:.2f} days")

```

```python
import pandas as pd
import numpy as np

# Load the calculated periods
input_file = 'cepheid_periods.csv'  # File with periods from previous step
periods_df = pd.read_csv(input_file)

# Coefficients for the Period-Luminosity Relation
a = -2.76  #  slope
b = -1.43  #  intercept

# Calculate the absolute magnitude
periods_df['logP'] = np.log10(periods_df['Period'])  # Logarithm of the period
periods_df['Absolute_Magnitude'] = a * periods_df['logP'] + b

print(periods_df[['ID', 'Period', 'Absolute_Magnitude']])
```

```python
import pandas as pd
import numpy as np

# Load the absolute magnitudes and periods
input_file = 'cepheid_absolute_magnitudes.csv'  # File with periods and absolute magnitudes
data = pd.read_csv(input_file)

# Load the original dataset
file_path = 'curves.csv'  # Replace with file path
curves_data = pd.read_csv(file_path)

curves_data.rename(columns={'id': 'ID', 'julian_date': 'JD', 'm': 'apparent_magnitude'}, inplace=True)

# Merge absolute magnitudes with apparent magnitudes
merged_data = data.merge(curves_data[['ID', 'apparent_magnitude']], on='ID')

# Calculating distances
merged_data['Distance_pc'] = 10 ** ((merged_data['apparent_magnitude'] - merged_data['Absolute_Magnitude'] + 5) / 5)

print(merged_data[['ID', 'Period', 'Absolute_Magnitude', 'apparent_magnitude', 'Distance_pc']])
```
