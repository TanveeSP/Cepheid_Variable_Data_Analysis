```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression


# Load the data
data = pd.read_csv('cepheids.csv')

# Extract the period and magnitude columns
P = data['Period']
M = data['M']

# Compute log10(P)
log_P = np.log10(P).values.reshape(-1, 1)

# Perform linear regression
model = LinearRegression()
model.fit(log_P, M)

# Extract the constants a and b
a = model.coef_[0]  # Slope
b = model.intercept_  # Intercept

print(f"a (slope): {a}")
print(f"b (intercept): {b}")
```

```python
import matplotlib.pyplot as plt

# Plot the original data
plt.scatter(P, M, label="Data", color='blue')

# Plot the fitted line
plt.plot(P, model.predict(np.log10(P).values.reshape(-1, 1)), label="Fitted Line", color='red')

# Labeling to axes
plt.xlabel('Period (P)')
plt.ylabel('Magnitude (M)')
plt.title('Cepheid Variables: Period vs Magnitude')

# Show legend
plt.legend()

# Display plot
plt.show()
```
